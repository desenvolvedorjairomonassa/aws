```python
import json
import boto3

bedrock = boto3.client(
    service_name='bedrock-runtime'
)

modelId = 'ai21.j2-ultra'
accept = 'application/json'
contentType = 'application/json'
prompt = """
Explain what democracy is to an 8 year old
"""
input = {
    'prompt': prompt, 
    'maxTokens': 1024,
    'temperature': 0.3,
    'topP': 1.0,
    'stopSequences': [],
    'countPenalty': {'scale': 0},
    'presencePenalty': {'scale': 0},
    'frequencyPenalty': {'scale': 0}
}
body=json.dumps(input)
response = bedrock.invoke_model(body=body, modelId=modelId, accept=accept,contentType=contentType)
response_body = json.loads(response.get('body').read())
print(json.dumps(response_body, indent=2))
```

```python
#response generated text without array 
completions = response_body['completions']
for part in completions:
    print(part['data']['text'])
```

```python
import json
import boto3

bedrock = boto3.client(
    service_name='bedrock-runtime'
)

modelId = 'mistral.mistral-7b-instruct-v0:2'
accept = 'application/json'
contentType = 'application/json'
prompt = """

Human: Explain what democracy is to an 8 year old.

Assistant:
"""
input = {
    "prompt": prompt,
    "max_tokens": 200,
    "temperature": 0.5,
}

body=json.dumps(input)
response = bedrock.invoke_model(body=body, modelId=modelId, accept=accept,contentType=contentType)
response_body = json.loads(response.get('body').read())
print(json.dumps(response_body, indent=2))
```

```python
generation = response_body.get("outputs")
for output in generation:
    print(output['text'])
```

Streaming API

```python
import json
import boto3

bedrock = boto3.client(
    service_name='bedrock-runtime'
)

modelId = 'mistral.mistral-7b-instruct-v0:2'
prompt = """

Human: explain DEMOCRACY to a university professor with 1000 words.

Assistant:
"""
input = {
    'prompt': prompt,
    'max_tokens': 1280, 
    'temperature': 1,
    'top_k': 1,
    'top_p': 1,
    'stop': []
}
body=json.dumps(input)
response = bedrock.invoke_model_with_response_stream(
    body=body, modelId=modelId, accept=accept,contentType=contentType)
stream = response.get('body')
if stream:
    for event in stream:
        chunk = event.get('chunk')
        if chunk:
            print(json.loads(chunk.get('bytes')))
```
